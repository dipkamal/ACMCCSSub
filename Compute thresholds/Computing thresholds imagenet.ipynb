```python

def compute_fpr(ben, t1, t2): 
    TN=0
    FN=0
    FP = 0 
    TP=0
    
    for value in ben: 
        if t1<=value<=t2:
            TN += 1
        else: 
            FP += 1
    
    return (FP/(FP+TN))*100
```

```python
t1=[1300,1800,2000]
t2=[5700,5700,5700]
for x, y in zip(t1,t2):
    print(compute_fpr(top_450_benign, x, y))
    print("---")
```
