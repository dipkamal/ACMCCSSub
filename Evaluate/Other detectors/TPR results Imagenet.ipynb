# Feature squeeze method with threshold computed from train set

```python
threshold = [1.36418, 1.3909, 1.299158]
```

```python
def compute_metrics(adv, threshold): 
    TN=0
    FN=0
    FP = 0 
    TP=0
    
    for value in adv: 
        if value>threshold:
            TP += 1
        else: 
            FN += 1
    
    
    return (TP/(TP+FN))*100
```

# approach 2 (a) and (b) combined

We combine the two approaches: 2(a) and 2(b). Since 2(a) works well for smaller epsilon while 2(b) works well for larger epsilon. So we can fix a detection strategy as: 
- If the image is assigned adversary by 2(a), its an adversary. 
- Else, if the approach 2(a) says its benign, most possibly it was an high epsilon attack, use approach 2(b) to check if its benign or adv. If it says, adv, its adv. Else, benign. 

```python
def compute_metrics(adv1, a, b, adv2, c, d): 
    TN=0
    FN=0
    FP = 0 
    TP=0
    
    for value1, value2 in zip(adv1, adv2): 
        if value1<a or value1>b:
            TP += 1
        else:
            if value2<c or value2>d:
                TP+=1
            else: 
                FN+=1
    
    return (TP/(TP+FN))*100
```

```python
def compute_metrics(adv1, a, b, adv2, c, d, adv3, e, f): 
    TN=0
    FN=0
    FP = 0 
    TP=0
    
    for value1, value2, value3 in zip(adv1, adv2, adv3): 
        if value1<a or value1>b:
            TP += 1
        else:
            if value2<c or value2>d:
                TP+=1
            else:
                if value3<e or value3>f:
                    TP+=1
                else:
                    FN+=1
    
    return (TP/(TP+FN))*100
```

```python
def compute_metrics(meanAb, a, b, medianAb, c,d, coefiqr, e, f, coefvar, g, h, iqr, i,j): 
    TN=0
    FN=0
    FP = 0 
    TP=0
    
    for value1, value2, value3, value4, value5 in zip(meanAb,medianAb,coefiqr,coefvar, iqr): 
        if value1<a or value1>b:
            TP += 1
        else:
            if value2<c or value2>d:
                TP +=1
            else: 
                if value3<e or value3>f:
                    TP +=1
                else:
                    if value4<g or value4>h:
                        TP +=1
                    else:
                        if value5<i or value5>j:
                            TP +=1
                        else:
                            FN+=1
    
    
    return (TP/(TP+FN))*100
```

```python
def compute_metrics(meanAb, a, b, medianAb, c,d, coefiqr, e, f, coefvar, g, h, ap1, i, j, ap2a, k, l, ap2b, m, n, iqr, o, p): 
    TN=0
    FN=0
    FP = 0 
    TP=0
    
    for value1, value2, value3, value4, value5, value6, value7, value8 in zip(meanAb,medianAb,coefiqr,coefvar,ap1,ap2a,ap2b, iqr): 
        if value1<a or value1>b:
            TP += 1
        else:
            if value2<c or value2>d:
                TP +=1
            else: 
                if value3<e or value3>f:
                    TP +=1
                else:
                    if value4<g or value4>h:
                        TP +=1
                    else: 
                        if value5<i or value5>j:
                            TP +=1
                        else: 
                            if value6<k or value6>l:
                                TP +=1
                            else:
                                if value7<m or value7>n:
                                    TP +=1
                                else:
                                    if value8<o or value8>p:
                                        TP +=1
                                    else:
                                        FN+=1
    
    
    return (TP/(TP+FN))*100
```
